```python
import nltk
import re
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

# 2 paragaraph worth of text as an input

```python
text = """ Artificial Intelligence (AI) is transforming the world rapidly. It helps machines learn from data, make decisions, and solve complex problems! Many industries such as healthcare, finance, and transportation are using AI technologies. However, proper data preparation is necessary before building any AI model.
Natural Language Processing (NLP) is a branch of AI. It allows computers to understand human language, analyze text, and generate responses. But raw text often contains punctuation marks, stop words, and other unnecessary elements; therefore preprocessing is important for better analysis."""
```

```python
print(text) # how the text look like in its original form
```

# Removing the punctuation from the text using regex

```python
text_no_punct = re.sub(r'[^\w\s]','',text)
print("\nText after removing punctuation:\n", text_no_punct)
```

```python
tokens = word_tokenize(text_no_punct) # tokenizing the texts
len(tokens)
```

# Removing the stopwords from the text

```python
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in tokens if word.lower() not in stop_words]
```

```python
print("\nAfter Stopword Removal:\n", filtered_words),print(len(filtered_words))
```

# performing stemming to the filtered_words

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
```

```python
print("\nAfter Stemming:\n", stemmed_words),print(len(stemmed_words))
```

# Performing the lemmatization to the text

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
```

```python

print("\nAfter Lemmatization:\n", lemmatized_words),print(len(lemmatized_words))
```
